```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
import graphviz
from subprocess import check_call
from IPython.display import Image
import global_funcs
```

```python
X_train = pd.read_csv('X_train.csv')
Y_train = pd.read_csv('Y_train.csv')
X_valid = pd.read_csv('X_valid.csv')
Y_valid = pd.read_csv('Y_valid.csv')
```

```python
def train(arbol_name,X,Y,X_test,Y_test,max_depth,cw = None):
    conf_string = "Arbol max_depg: "+str(max_depth)+' class_wigth: '+str(cw)
    arbol = DecisionTreeClassifier(max_depth=max_depth,class_weight=cw)
    arbol.fit(X,Y)
    y_pred_test = arbol.predict(X_test)
    y_pred_train = arbol.predict(X)
    report = global_funcs.getReport(Y_test["passenger_survived"],Y["passenger_survived"],y_pred_test, y_pred_train)
    df_report = pd.DataFrame(report)
    
    y_names = ["No sobrevivio","Sobrevivio"]
    x_names = X.columns.values.tolist()
    
    export_graphviz(arbol,out_file=arbol_name,feature_names=x_names,class_names=y_names,impurity=False,filled=True)
    
    return {
        "arbol":arbol,
        "reporte":df_report,
        "conf_string":conf_string
    }   
```

```python
max_depts = [None,3,4,5,6]
class_weights = [None,'balanced']

modelos = []

for c in class_weights:
    for p in max_depts:
        modelo = train("modelo.dot",X_train,Y_train,X_valid,Y_valid,p,c)#entrenando el modelo
        global_funcs.writeInExcel(modelo['reporte'],"bitacora_arbol.xlsx",modelo['conf_string'])#escribiendo en bitacora excel
        modelo["reporte"]
        modelos.append({
                    "accuracy": modelo["reporte"].iloc[0]["metrics_test"],
                    "presicion":modelo["reporte"].iloc[1]["metrics_test"],
                    "recall": modelo["reporte"].iloc[2]["metrics_test"],
                    "f1_score": modelo["reporte"].iloc[3]["metrics_test"],
                    "max_depth":p,
                    "class_weith":c})
```
