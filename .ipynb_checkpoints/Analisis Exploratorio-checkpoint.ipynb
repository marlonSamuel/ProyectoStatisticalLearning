```python
def contingenceTable(df,column_a, column_b):
    contingence = pd.crosstab(df[column_a],df[column_b])
    contingence.plot(kind="bar")
    plt.title("Gráfica de sobrevivientes y no sobrevivientes de "+column_a)
    plt.xlabel(column_a)
    plt.ylabel("Sobrevivientes")
    plt.show()
    
    hor = contingence.astype("float").div(contingence.sum(axis=1),axis=0)
    ver = contingence.astype("float").div(contingence.sum(axis=0),axis=1)
    columns_a = df[column_a].dropna().unique()
    return pd.DataFrame({
                         "sobrevivio":contingence[1],
                         "no sobrevivio":contingence[0],
                        "% sobrevivio": hor[1],
                        "% no sobrevivio": hor[0],
                        "% general sobrevivio": ver[1],
                        "% general no sobrevivio": ver[0]})
```

```python
#tabla de contingencia para genero
c_s = contingenceTable(data,"passenger_sex","passenger_survived")
c_s
```

* Hombres sobrevivieron: 109
* Hombres murieron: 468
* Mujeres sobrevieron: 233
* Mujeres muerieron: 81

Se observa un dato interesante, y esque el porcentaje de hombres (entre todos los hombres) que no sobrevivieron es alto 81%, por el contrario, en las mujeres (entre todas las mujeres) el porcentaje es de 25%

a simple vista los datos nos indica que:
* hay una alta probabilidad de **no sobrevivir** en caso de ser hombre.
* Hay una alta probabilidad de **sobrevivir** en caso de ser mujer.

Tambien se observa en la gráfica que el porcentaje de los hombres vs las mujeres a bordo murieron es del 68% de las mujeres murieron vs 31.87% de hombre.

Se podría decir que el genero es una caractaristica significante para predecir si sobrevive o no una persona.
