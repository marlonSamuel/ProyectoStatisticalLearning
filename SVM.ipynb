```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn import svm
import global_funcs
from joblib import dump
```

```python
X_train = pd.read_csv('X_train.csv')
Y_train = pd.read_csv('Y_train.csv')
X_valid = pd.read_csv('X_valid.csv')
Y_valid = pd.read_csv('Y_valid.csv')
```

```python
#función de entrenamiento
def train(X,Y,X_test,Y_test,kernel,cv=None,deg=1):
    conf_string = 'svm_kernel='+kernel+'_class_weitht='+str(cv)+("_degree="+str(deg) if deg > 1 else "")
    clf = svm.SVC(kernel=kernel,class_weight=cv,degree=deg)
    clf.fit(X_train, Y_train.values.ravel())
    y_pred_test = clf.predict(X_test)
    y_pred_train = clf.predict(X)
    report = global_funcs.getReport(Y_test["passenger_survived"],Y["passenger_survived"],y_pred_test, y_pred_train)
    df_report = pd.DataFrame(report)
    
    return {
        "clf":clf,
        "reporte":df_report,
        "conf_string":conf_string
    }
```

### Experimentación

Se experimentará con diferentes hyperparametros para encontrar el mejor modelo.

```python
kernels = ['rbf','linear','poly','sigmoid']
degrees = [2,3,4,5]
class_weights = [None,'balanced']

modelos = []

for k in kernels:
    for c in class_weights:
        if k == 'poly':
            for d in degrees:
                modelo = train(X_train,Y_train,X_valid,Y_valid,k,c,d)#entrenando el modelo
                global_funcs.writeInExcel(modelo['reporte'],"bitacora_svc.xlsx",modelo['conf_string'])#escribiendo en bitacora excel
                modelo["reporte"]
                modelos.append({
                    "accuracy": modelo["reporte"].iloc[0]["metrics_test"],
                    "presicion":modelo["reporte"].iloc[1]["metrics_test"],
                    "recall": modelo["reporte"].iloc[2]["metrics_test"],
                    "f1_score": modelo["reporte"].iloc[3]["metrics_test"],
                    "kernel":k,
                    "class_weith":c,
                    "degree ":d})
        modelo = train(X_train,Y_train,X_valid,Y_valid,k,c)#entrenando el modelo
        global_funcs.writeInExcel(modelo['reporte'],"bitacora_svc.xlsx",modelo['conf_string'])#escribiendo en bitacora excel
        modelo["reporte"]
        modelos.append({
                    "accuracy": modelo["reporte"].iloc[0]["metrics_test"],
                    "presicion":modelo["reporte"].iloc[1]["metrics_test"],
                    "recall": modelo["reporte"].iloc[2]["metrics_test"],
                    "f1_score": modelo["reporte"].iloc[3]["metrics_test"],
                    "kernel":k,
                    "class_weith":c})
```

```python
#resumen de modelos
resumen = pd.DataFrame(modelos)
resumen
```
